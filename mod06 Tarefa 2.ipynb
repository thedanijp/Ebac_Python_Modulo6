```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
brsus = pd.read_csv('sinasc_RO_2019.csv')
brsus.head()
```

```python
# Boxplot da idade da mãe com personalização
plt.figure(figsize=(10, 6))
sns.boxplot(x='IDADEMAE', data=brsus)
plt.title('Distribuição da Idade das Mães')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Boxplot do peso do bebê com personalização
plt.figure(figsize=(10, 6))
sns.boxplot(x='PESO', data=brsus)  # Assumindo que a coluna do peso é 'PESO'
plt.title('Distribuição do Peso dos Bebês')
plt.xlabel('Peso do Bebê (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Histograma da idade da mãe com densidade
plt.figure(figsize=(10, 6))
sns.histplot(data=brsus, x='IDADEMAE', kde=True)
plt.title('Distribuição da Idade das Mães')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Histograma do peso do bebê com densidade
plt.figure(figsize=(10, 6))
sns.histplot(data=brsus, x='PESO', kde=True)
plt.title('Distribuição do Peso dos Bebês')
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('Frequência')
plt.show()
```

```python
# Detectando outliers na idade da mãe
q1 = brsus['IDADEMAE'].quantile(0.25)
q3 = brsus['IDADEMAE'].quantile(0.75)
iqr = q3 - q1
outliers = brsus[(brsus['IDADEMAE'] < q1 - 1.5 * iqr) | (brsus['IDADEMAE'] > q3 + 1.5 * iqr)]
print("Número de outliers na idade da mãe:", len(outliers))
```

```python
# Correlação entre idade da mãe e peso do bebê (se houver outras variáveis relevantes)
sns.scatterplot(x='IDADEMAE', y='PESO', data=brsus)
plt.title('Correlação entre Idade da Mãe e Peso do Bebê')
plt.xlabel('Idade da Mãe')
plt.ylabel('Peso do Bebê (g)')
plt.show()
```
